Q1. What is a database? Differentiate between SQL and NoSQL databases.

A database is an organized collection of data that allows easy storage, retrieval, and management of information.

Difference between SQL and NoSQL databases
Feature	SQL Database	NoSQL Database
Data model	Relational (tables, rows, columns)	Non-relational (documents, key-value, etc.)
Schema	Fixed schema	Flexible / dynamic schema
Query language	SQL	Different APIs / query styles
Best for	Structured data	Semi-structured / unstructured data
Scalability	Vertical scaling (mostly)	Horizontal scaling

Examples

SQL databases: MySQL, PostgreSQL

NoSQL databases: MongoDB

Q2. What is DDL? Explain CREATE, DROP, ALTER and TRUNCATE with examples.

DDL (Data Definition Language) is used to define and change the structure of database objects.

➤ CREATE

Used to create a table.

```python
CREATE TABLE student(
    id INT,
    name VARCHAR(50)
);
```

➤ DROP

Used to permanently remove a table.

DROP TABLE student;
➤ ALTER

Used to modify an existing table.

ALTER TABLE student ADD age INT;
➤ TRUNCATE

Used to remove all records from a table (structure remains).

TRUNCATE TABLE student;

Q3. What is DML? Explain INSERT, UPDATE and DELETE with examples.

DML (Data Manipulation Language) is used to manipulate the data stored in tables.

➤ INSERT
INSERT INTO student VALUES (1,'Amit',20);
➤ UPDATE
UPDATE student SET age = 21 WHERE id = 1;
➤ DELETE
DELETE FROM student WHERE id = 1;
Q4. What is DQL? Explain SELECT with an example.

DQL (Data Query Language) is used to retrieve data from the database.

The main command is SELECT.

SELECT * FROM student;

This displays all records from the table.

Q5. Explain Primary Key and Foreign Key.
✅ Primary Key

Uniquely identifies each record in a table.

Cannot be NULL.

No duplicate values.

Example:

id INT PRIMARY KEY
✅ Foreign Key

A key in one table that refers to the primary key of another table.

Maintains relationship between tables.

Example:

FOREIGN KEY (dept_id) REFERENCES department(dept_id)

Q6. Write a Python code to connect MySQL to Python.

Explain cursor() and execute().**

```python
import mysql.connector

con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="password",
    database="school"
)

cur = con.cursor()

cur.execute("SELECT * FROM student")

for row in cur:
    print(row)

con.close()
```

➤ cursor()

Creates a cursor object.

It is used to execute SQL commands and fetch records.

➤ execute()

Used to execute an SQL query using the cursor.

Example:

cur.execute("SELECT * FROM student")
